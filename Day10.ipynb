```python
dj_open_ts = pd.Series(dow_jones_raw\
                       .query('stock == "AA"')
                       [
        ['date', 'open'] ].set_index('date')['open']\
                       .replace('\$', '', regex=True).astype(float) )
dj_open_ts
```

```python
train_ts = dj_open_ts.iloc[:-5]
test_ts = dj_open_ts.iloc[-5:]
print(train_ts.shape, test_ts.shape)
test_ts
```

```python
train_ts.plot()
```

```python
test_ts.plot()
```

```python
# ARIMA example
from statsmodels.tsa.arima_model import ARIMA
```

```python
model = ARIMA(train_ts, order=(2, 1, 2))
model_fit = model.fit(disp=False)
```

```python
y_predicted = model_fit.predict(test_ts.index[0],
                         test_ts.index[-1],
                         typ='levels')
print(y_predicted)
```

## Multivariate timeseries

```python
dow_jones_raw
```

```python
dj_multivar_ts = dow_jones_raw.query('stock == "AA"')[ ['date', 'open'] ].set_index('date')
dj_multivar_ts['AA'] = dj_multivar_ts['open'].replace('[\$,]', '', regex=True).astype(float)
dj_multivar_ts = dj_multivar_ts.drop('open', axis =1)
dj_multivar_ts['T'] = dow_jones_raw.query('stock == "T"')[ ['date', 'open'] ].set_index('date').replace('[\$,]', '', regex=True).astype(float)
dj_multivar_ts
```

```python
dow_jones_raw.pivot(index = 'date', columns = 'stock', values='open')
```

```python
dow_jones_raw.pivot(index = 'date', columns = 'stock', values='open').shape
```

```python
dow_jones_raw.pivot(index = 'date', columns = 'stock', values='open')[ ['AA', 'T']]
```

```python
train_ts = dj_multivar_ts.iloc[:-5]
test_ts = dj_multivar_ts.iloc[-5:]
print(train_ts.shape, test_ts.shape)
test_ts
```

```python
# VAR example
from statsmodels.tsa.vector_ar.var_model import VAR
```

```python

# fit model
model = VAR(train_ts)
model_fit = model.fit()

# make prediction
y_predicted = model_fit.forecast(train_ts[ ['AA','T'] ].as_matrix() , steps=5)
print(y_predicted)
```

```python
# VARMA example
from statsmodels.tsa.statespace.varmax import VARMAX
```

```python

# fit model
model = VARMAX(train_ts, order=(1, 1))
model_fit = model.fit(disp=False)
# make prediction
y_predicted = model_fit.forecast(steps = 5)
print(y_predicted)
```
